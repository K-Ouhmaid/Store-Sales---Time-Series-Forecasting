```python
data = pd.read_csv("https://qwerty159.com/data.csv")
```

```python
data.shape
```

```python
data.tail()
```

```python
#rename columns 
data = data.rename(columns={'Shipment ID' : 'shipment_id','Box Type Original' : 'box_original','Box Type Final' : 'box_final','Pick Date' : 'pick_date','Pick time' : 'pick_time','Pack Date' : 'pack_date','Pack Time' : 'pack_time','Pack Date.1' : 'pack_date_time'}) 
```

```python
data
```

```python
#checking types
data.dtypes
```

```python
#transfer object to datatime type

data['pick_date_time']=data['pick_date'].apply(lambda x :str(x)) + ' ' + data.pick_time

```

```python
#remove redudndant informations

data=data.drop(['pick_date', 'pick_time','pack_date','pack_time'], axis=1)
```

```python
data.head(5)
```

```python
#make sure the pick & pack datetime are in the same format

data["pack_date_time"]=pd.to_datetime(data["pack_date_time"], errors='coerce')
data["pick_date_time"]=pd.to_datetime(data["pick_date_time"], errors='coerce')
data.pack_date_time=pd.to_datetime(data['pack_date_time'], format='%Y-%m-%d %h:%m:%s ')
data.pick_date_time=pd.to_datetime(data['pick_date_time'], format='%Y-%m-%d %h:%m:%s ')

```

```python
data.dtypes
```

```python
#checking nulls

data.isnull().sum()
```

```python
#time diff between pick & pack for each shipment

data['pick_pack_time']=round((data.pack_date_time - data.pick_date_time).dt.seconds/60,2)
```

```python
data
```

```python
data.describe()
```

```python
#are we always the same orginal box

data['same_box']=data[['box_original','box_final']].apply(lambda x: True if x['box_original']==x['box_final'] else False, axis=1)

```

```python
#are we always shiping 1 piece

data['pieces_over_1']=data[['Pieces']].apply(lambda x: True if x['Pieces']>1 else False, axis=1)
```

```python
def f(x):
    if (x > 4) and (x <= 8):
        return 'Early Morning'
    elif (x > 8) and (x <= 12 ):
        return 'Morning'
    elif (x > 12) and (x <= 16):
        return'Noon'
    elif (x > 16) and (x <= 20) :
        return 'Eve'
    elif (x > 20) and (x <= 24):
        return'Night'
    elif (x <= 4):
        return'Late Night'
```

```python
#in which part of the day the shipement is packed

data['session'] = data['pack_date_time'].dt.hour.apply(f)
```

```python
data
```

```python
data_type=data.Type.value_counts().reset_index().rename(columns={'index' : 'type','Type' : 'count'}) 


```

```python
data_type['percentage']=round(data_type["count"]/data.shape[0]*100,3)
```

```python
data_type
```

```python
temp=data.groupby(["Type","Status"]).agg({"shipment_id" :np.size}).reset_index().rename(columns={'shipment_id' : 'count','Type' : 'type'}) 
```

```python
temp
```

```python
temp=temp.merge(data_type,on="type",how="left").rename(columns={'count_x' : 'countStatus','count_y' : 'countType','percentage' : 'percentageType'}) 
```

```python
temp
```

```python
temp["percentageStatus"]=round(temp.countStatus/temp.countType*100,2)
```
